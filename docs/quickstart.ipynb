```python
import pandas as pd

from lactopy import OBLA
```

```python
df = pd.read_csv('../notebooks/data/test.csv')
```

```python
# fit model returns a scikit learn like model object
obla = OBLA()
obla.fit(df['watt'], df['lactate'], method='spline')
```

```python
# plot fit of data
obla.plot.plot_fit()
```

```python
# predict intensity at 4 mmol
obla.predict(4)
```

```python
# Plot the prediction at 4 mmol
obla.plot.plot_predictions(4)
```

```python
from lactopy.lactate_models import Dmax
```

```python
model = Dmax()
model.fit(df['watt'], df['lactate'], impl='modified')
# plot fit of data
model.plot.plot_fit()
```

```python
model.plot.plot_predictions()
```

```python
from lactopy.lactate_models.lt1_si import LT1_si
```

```python
model = LT1_si()
model.fit(df['watt'], df['lactate'], si=0.5)
lt1_intensity = model.predict()
print(f"LT1 occurs at: {lt1_intensity:.2f} watts")
```

```python
model.plot.plot_predictions()
```
