```python
import pandas as pd

from lactopy.lactate_models.OBLA import OBLA
from lactopy.lactate_models.dmax import Dmax
```

```python
df = pd.read_csv('../tests/test_data/lactate_test_data copy.csv')
```

```python
# fit model returns a scikit learn like model object
obla = OBLA()
obla.fit(df['watt'], df['lactate'])
```

```python
# plot fit of data
obla.plot.plot_fit()
```

```python
# predict intensity at 4 mmol
obla.predict(4)
```

```python
# Plot the prediction at 4 mmol
obla.plot.plot_predictions(4)
```

```python
dmax_model = Dmax()
dmax_model.fit(df['watt'], df['lactate'])
dmax_model.plot.plot_predictions()
```
